# Introduction to classification analysis

The classification analysis is used when we want to predict a categorical variable. This implies in attributing each of your samples to a "category" or "class".

```python
%matplotlib inline
```

```python
import pandas as pd
import seaborn as sns ; sns.set()
import matplotlib.pyplot as plt
```

```python
debts= pd.read_csv('dados/dividas.tsv', sep = '\t')
debts.head()
```

```python
fig, ax = plt.subplots(figsize=(16, 8))
sns.scatterplot(x='balanco', y='renda', hue='debito', data=debts, alpha=0.5, ax=ax)
```

```python
fig, axs = plt.subplots(1,2, figsize=(16, 8))
sns.boxplot(x="debito", y="balanco", data=debts, ax=axs[0])
sns.boxplot(x="debito", y="renda", data=debts, ax=axs[1])
```
